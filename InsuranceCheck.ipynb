# One Hot Encoding Using Loop

```python
df7.head()
```

```python
cols = ['gender']
for col in cols:
    dummy = pd.get_dummies(df7[col], drop_first=True, prefix='gender')
    df7 = pd.concat([df7, dummy], axis=1).drop(col, axis=1)

cols = ['smoker']
for col in cols:
    dummy1 = pd.get_dummies(df7[col], drop_first=True, prefix='smoker')
    df7 = pd.concat([df7, dummy1], axis=1).drop(col, axis=1)

cols = ['region']
for col in cols:
    dummy2 = pd.get_dummies(df7[col], drop_first=True, prefix='region')
    df7 = pd.concat([df7, dummy2], axis=1).drop(col, axis=1)

df7.head()
```

# Ordinal Encoding

```python
df8.head()
```

```python
Gender = df8.gender.unique()
Gender
```

```python
Smoker = df8.smoker.unique()
Smoker
```

```python
Region = df8.region.unique()
Region
```

```python
from sklearn.preprocessing import OrdinalEncoder
od = OrdinalEncoder(categories=[Gender])
od1 = OrdinalEncoder(categories=[Smoker])
od2 = OrdinalEncoder(categories=[Region])
```

```python
encoded_values = od.fit_transform(df8[['gender']])
encoded_values
```

```python
encoded_values1 = od1.fit_transform(df8[['smoker']])
encoded_values1
```

```python
encoded_values2 = od2.fit_transform(df8[['region']])
encoded_values2
```

```python
encoded_values = pd.DataFrame(encoded_values,columns=['gender'])
encoded_values.head()
```

```python
encoded_values1 = pd.DataFrame(encoded_values1,columns=['smoker'])
encoded_values1.head()
```

```python
encoded_values2 = pd.DataFrame(encoded_values2,columns=['region'])
encoded_values2.head()
```

```python
df8 = pd.concat([df8.drop(['gender', 'smoker', 'region'], axis=1), encoded_values, encoded_values1, encoded_values2], axis=1)
df8.head()
```

# Ordinal Encoding Using loop

```python
df9.head()
```

```python
df_encoded = df9.copy()
```

```python
for column in df_encoded.columns:
    if df_encoded[column].dtype == 'object':
        unique_values = df_encoded[column].unique()
        od = OrdinalEncoder(categories=[unique_values])
        encoded_values = od.fit_transform(df_encoded[[column]])
        encoded_df = pd.DataFrame(encoded_values, columns=[column])
        df_encoded[column] = encoded_df[column]
df9 = df_encoded
```
